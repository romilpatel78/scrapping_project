```python
import pandas as pd
```

```python
#df = pd.read_csv("C:/Users/Owner/Desktop/Scrapping Project/data.csv")
df = pd.read_csv("C:/Users/Owner/Documents/salary_project/salary_proj/glassdoor.csv")
```

```python
df
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
df = df[df['Salary Estimate'] !='-1']
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
remove_dlr_K = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
hr_min = remove_dlr_K.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
df['salary_min'] = hr_min.apply(lambda x: int(x.split('-')[0]))
```

```python
df['salary_max'] = hr_min.apply(lambda x: int(x.split('-')[1]))
```

```python
df['avg_salary'] = (df.salary_min+df.salary_max)/2
```

```python
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-3], axis = 1)
```

```python
#df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
#df.job_state.value_counts()
```

```python
df['some_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
```

```python
df['age'] = df.Founded.apply(lambda x: x if x<1 else 2020 - x)
```

```python
df['python_ym'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df.python_ym.value_counts()
```

```python
df['r_ym'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() else 0)
df.r_ym.value_counts()
```

```python
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark.value_counts()
```

```python
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws.value_counts()
```

```python
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel.value_counts()
```
